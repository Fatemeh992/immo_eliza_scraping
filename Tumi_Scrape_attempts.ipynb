```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time
import csv

# Setup WebDriver (assuming Chrome here)
driver = webdriver.Chrome()  # Ensure that chromedriver is in your PATH

# Open the real estate website
url = 'https://www.immoweb.be/en/search/apartment/for-sale'
driver.get(url)

# List to store property data
properties = []

try:
    # Loop through pages (adjust this range or logic according to the pagination)
    for page in range(1, 5):  # Example for 4 pages; change as necessary
        print(f"Scraping page {page}")
        
        # Find all property links on the current page
        property_elements = driver.find_elements(By.XPATH, '//*[@id="main-content"]/div/li[5]')  
        
        for prop in property_elements:
            link = prop.get_attribute("href")
            driver.get(link)  # Open each property link
            
            # Extract property details
            try:
                title = driver.find_element(By.CSS_SELECTOR, "h1.title-class").text  # Adjust selector
                price = driver.find_element(By.CSS_SELECTOR, "span.price-class").text  # Adjust selector
                location = driver.find_element(By.CSS_SELECTOR, "div.location-class").text  # Adjust selector
                description = driver.find_element(By.CSS_SELECTOR, "p.description-class").text  # Adjust selector
                
                # Append the data to the list
                properties.append({
                    "title": title,
                    "price": price,
                    "location": location,
                    "description": description,
                    "link": link
                })
                
            except Exception as e:
                print(f"Error extracting data for {link}: {e}")
            
            # Go back to main page to continue to next property
            driver.back()
            time.sleep(2)  # Delay to avoid overloading the server

        # Click on the next page button if exists
        try:
            next_button = driver.find_element(By.CSS_SELECTOR, "button.next-page-class")  # Adjust selector
            ActionChains(driver).move_to_element(next_button).click().perform()
            time.sleep(2)  # Wait for the page to load
        except Exception as e:
            print("No more pages or next button not found.")
            break

finally:
    driver.quit()

# Save data to CSV
with open("properties.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["title", "price", "location", "description", "link"])
    writer.writeheader()
    writer.writerows(properties)

print("Scraping completed and data saved to properties.csv")
```
